```python
# Fill NaN values with a default value (e.g., 0)
natural_disas.fillna(0, inplace=True)

# Convert columns to integers
natural_disas1 = natural_disas.astype({"Start Month": "int", "Start Day": "int"})

print(natural_disas1.dtypes)
```

```python
natural_disas1 = natural_disas.astype({"End Month":"int64","End Day":"int64"})
print(natural_disas1.dtypes)
```

```python
cols=["Start Year","Start Month","Start Day"]
natural_disas1['Start Date'] = natural_disas1[cols].apply(lambda x: '-'.join(x.values.astype(str)), axis="columns")
```

```python
cols=["End Year","End Month","End Day"]
natural_disas1['End Date'] = natural_disas1[cols].apply(lambda x: '-'.join(x.values.astype(str)), axis="columns")
```

```python
columns_to_delete = ['Start Year','Start Month', 'Start Day', 'End Year','End Month','End Day']
natural_disas1.drop(columns=columns_to_delete, inplace=True)
```

```python
c=natural_disas1.corr()
sns.heatmap(c, cmap='tab10', linewidths='0.05', annot=True)
plt.show()
```

```python
# Assuming columns named 'Disaster_Type' (or similar) and 'Fatalities'
disaster_counts = natural_disas1['Disaster Type'].value_counts()  # Count disasters by type
total_fatalities = natural_disas1['Total Deaths'].sum()  # Total lives lost

# Create the figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(50,15))

# Subplot 1: Disaster Count Bar Plot
ax1.bar(disaster_counts.index, disaster_counts.values)
ax1.set_xlabel("Disaster Type")
ax1.set_ylabel("Number of Disasters")
ax1.set_title("Number of Disasters by Type")
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for readability


# Subplot 2: Total Fatalities Bar Plot
ax2.bar(["Total Death"], [total_fatalities])
ax2.set_xlabel("Total Damages ('000 US$)")
ax2.set_ylabel("Total Affected")
ax2.set_title("Total Death")

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
